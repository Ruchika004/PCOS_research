# Hypothesis Testing

```python
table = pd.crosstab(df['Hormonal_Imbalance'],df['PCOS'])
table
```

```python
from scipy.stats import chi2_contingency
```

```python
#Hormonal_Imbalance and PCOS

chi2, p, dof, ex = chi2_contingency(table)
print(f"Chi-square Test Result (p-value): chi2 = {chi2}, p-value = {p}")
```
